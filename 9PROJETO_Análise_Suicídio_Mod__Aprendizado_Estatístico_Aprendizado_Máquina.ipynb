#<font color=yellow>Tendo por objetivo deixar os dados mais limpos poss√≠veis, ser√° eliminado todas as linhas nulas do conjunto de dados (DataSet)

```python
#Eliminando qualquer linha nula

data = data.dropna()
data.shape
```

```python
#verificando se h√° outlier e removendo-o

data=data.sort_values(by=["suicide_rate"])
q1=data["suicide_rate"].quantile(0.25)
q3=data["suicide_rate"].quantile(0.75)
iqr=q3-q1
lwo=q1-1.5*iqr
upo=q3+1.5*iqr
data=data[(data.suicide_rate<upo)&(data.suicide_rate>lwo)]
data=data.sort_index().reset_index(drop=True)
data.shape
```

#<font color=yellow>Criando uma c√≥pia do conjunto de dados para teste estat√≠stico

```python
#criando uma c√≥pia do conjunto de dados para teste estat√≠stico
stat_data = data.copy()
stat_data
```

#<font color=yellow>As colunas rotuladas n√£o num√©ricas como: pa√≠s, ano, sexo, faixa et√°ria e gera√ß√£o devem ser convertidas em r√≥tulos num√©ricos que podem ser feitos usando o LabelEncoder do SkLearn.

```python
#codificando os recursos categ√≥ricos com LabelEncoder

from sklearn.preprocessing import LabelEncoder
import pickle

# Crie um exemplo de DataFrame 'stat_data' (substitua isso pelos seus pr√≥prios dados)
# Certifique-se de que 'stat_data' seja um DataFrame v√°lido com as colunas especificadas em 'categorical'.
# Exemplo:
# stat_data = pd.DataFrame({'country': ['USA', 'Canada', 'UK'], 'year': [2020, 2021, 2022], ...})

# Certifique-se de ter definido seu DataFrame 'stat_data' com os dados reais aqui

categorical = ['country', 'year', 'age_group', 'gender', 'generation']
label_encoders_dict = {}

for column in categorical:
    le = LabelEncoder()
    le.fit(stat_data[column])
    label_encoders_dict[column] = le
    stat_data[column] = le.transform(stat_data[column])

# Salve os LabelEncoders em um arquivo pickle
with open('/content/label_encoders.pkl', 'wb') as file:
    pickle.dump(label_encoders_dict, file)
```

```python
#abrindo arquivo

file = open('/content/label_encoders.pkl', "rb")
label = pickle.load(file)
label
```

#<font color=yellow>A padroniza√ß√£o de um conjunto de dados √© recomendado para modelos de aprendizado estat√≠stico no aprendizado de m√°quina.
```python
#descri√ß√£o do conjunto de dados

stat_data.describe().T
```

```python
#Dimensionando as colunas de dados num√©ricos com RobustScalar

numerical = ['suicide_count', 'population', 'gdp_for_year','gdp_per_capita']

from sklearn.preprocessing import RobustScaler

rc = RobustScaler()
stat_data[numerical] = rc.fit_transform(stat_data[numerical])
```

```python
import pickle
pickle.dump(rc, open('/content/label_encoders.pkl', "wb"))
```

```python
# Separando e atribuindo recursos e colunas de destino a X e y

y = stat_data['suicide_rate']
X = stat_data.drop('suicide_rate',axis=1)
X.shape, y.shape
```

##<font color=khaki>Testes estat√≠sticos s√£o usados em testes de hip√≥teses. Eles podem ser usados para: determinar se uma vari√°vel preditora tem uma rela√ß√£o estatisticamente significativa com uma vari√°vel de resultado e estimar a diferen√ßa entre dois ou mais grupos.

#<font color=yellow>H0:

##<font color=khaki>N√£o h√° diferen√ßa nas taxas de suic√≠dio entre homens e mulheres (Nulo).

#<font color=yellow>H1:

##<font color=khaki>H√° diferen√ßa nas taxas de suic√≠dio entre homens e mulheres (Alternativo)

#<font color=yellow>Conclus√£o do teste:

##<font color=khaki>Ao realizar o teste T, o resultado obtido √© a rejei√ß√£o da hip√≥tese nula. Isso basicamente significa que h√° diferentes taxas de suic√≠dio entre homens e mulheres.

#<font color=yellow>Teste 2:

##<font color=khaki>Descobrir a depend√™ncia da taxa de suic√≠dio com a idade.

##<font color=khaki>Descobrir se existe depend√™ncia da taxa de suic√≠dio em rela√ß√£o √† idade por meio do teste Qui-Quadrado. As declara√ß√µes de hip√≥teses para este teste s√£o:

#<font color=yellow>H0:

##<font color=khaki>Taxa de suic√≠dio e idade s√£o independentes (Nula).

#<font color=yellow>H1:

##<font color=khaki>A taxa de suic√≠dio e a idade s√£o dependentes (Alternativo).

```python
#Criando Tabela de Conting√™ncia
contingency_table = pd.crosstab(stat_data.suicide_rate, stat_data.age_group)
```

```python
#N√≠vel de signific√¢ncia 5%
alpha=0.05
```

```python
chistat, p, dof, expected = stats.chi2_contingency(contingency_table )
```

```python
#valor cr√≠tico
critical_value=stats.chi2.ppf(q=1-alpha,df=dof)
print('critical_value:',critical_value)
```

```python
print('Significance level: ',alpha) #n√≠vel de signific√¢ncia
print('Degree of Freedom: ',dof) #grau de liberdade
print('chi-square statistic:',chistat) #estat√≠stica qui-quadrado
print('critical_value:',critical_value) #valor cr√≠tico
print('p-value:',p) #valor-p
```

#<font color=yellow>Aqui, o valor-p = 0,0 e um valor-p baixo sugerem que sua amostra fornece evid√™ncias suficientes de que voc√™ pode rejeitar H0 para toda a popula√ß√£o.

```python
#compare estat√≠stica_qui_quadrado com valor_cr√≠tico e valor-p, que √© o
#probabilidade de obter qui-quadrado> 0,09 (estat√≠stica_qui_quadrado)
if chistat>=critical_value:
    print("Reject H0,There is a dependency between Age group & Suicide rate.") #Rejeite H0, h√° uma depend√™ncia entre faixa et√°ria e taxa de suic√≠dio
else:
    print("Retain H0,There is no relationship between Age group & Suicide rate.") #Retem H0, n√£o h√° rela√ß√£o entre faixa et√°ria e taxa de suic√≠dio

if p<=alpha:
    print("Reject H0,There is a dependency between Age group & Suicide rate.") #Rejeite H0, h√° uma depend√™ncia entre faixa et√°ria e taxa de suic√≠dio
else:
    print("Retain H0,There is no relationship between Age group & Suicide rate.") #Retem H0, n√£o h√° rela√ß√£o entre faixa et√°ria e taxa de suic√≠dio
```
